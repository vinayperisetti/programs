```python
import matplotlib.pyplot as plt

Country = ['USA','INDIA','GERMANY','UK']
GDP_Per_Capita = [1000,2000,3000,4000]

plt.bar(Country,GDP_Per_Capita)
plt.title("Country vs GDP per Capita")
plt.xlabel('Country')
plt.ylabel('GDP_Per_Capita')
plt.suptitle("INDIA")
plt.show()
```

```python
import matplotlib.pyplot as plt

Country = ['USA','INDIA','GERMANY','UK']
GDP_Per_Capita = [1000,2000,3000,4000]

plt.barh(Country,GDP_Per_Capita)
plt.title("Country vs GDP per Capita")
plt.xlabel('Country')
plt.ylabel('GDP_Per_Capita')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35,50,15])

plt.pie(y)
plt.title("pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
y = np.array([35,50,15])
co=["r","b","yellow"]
plt.pie(y,colors=co,labels=["dsp","cd","coa"])
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(1,2,1)
plt.plot(x,y)
plt.title("linee")

#plot2
x= np.array([0,1,2,3])
y = np.array([10,20,40,50])

plt.subplot(1,2,2)
plt.pie(y)
plt.title("pieeee")
plt.suptitle("subplot")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

#plot1
x = np.array([0,1,2,3])
y = np.array([3,8,1,10])

plt.subplot(4,1,1)
plt.plot(x,y)
plt.title("linee")

#plot2
x= np.array([0,1,2,3])
y = np.array([10,20,40,50])

plt.subplot(4,1,2)
plt.pie(y)
plt.title("pieeee")

#plot3
x = np.array([0,1,2,3])
y = np.array([1000,2000,3000,4000])

plt.subplot(4,1,3)
plt.bar(x,y,color="r")
plt.title("plot3")
plt.xlabel('x')
plt.ylabel('y')

#plot4
x = np.array([0,1,2,3])
y = np.array([1000,2000,3000,4000])

plt.subplot(4,1,4)
plt.barh(x,y,color="r")
plt.title("plot3")
plt.xlabel('x')
plt.ylabel('y')
plt.suptitle("subplot")
plt.show()
```
